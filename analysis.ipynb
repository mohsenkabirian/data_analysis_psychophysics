```python
import pandas as pd
import os
from glob import glob  # Explicitly import glob

# Get current directory path
path = os.getcwd()

# Find all CSV files
csv_files = glob(os.path.join(path, "*.csv"))


# Read the Excel file with header row
df_all_old = pd.read_excel('simon.xlsx', header=0)

df_all_new = df_all_old


# Process each CSV file
for file in csv_files:
  
  # Read the CSV file
  df = pd.read_csv(file)
  # accuracy_column = df['key_resp_trial.corr']
  # df_filtered = df[df['key_resp_trial.rt'] > 0.1]

  mean = df['key_resp_trial.rt'].mean()
  std = df['key_resp_trial.rt'].std()

  df_filtered = df[df['key_resp_trial.keys'] != 'None']
  df_filtered = df_filtered.query('0.1 < `key_resp_trial.rt` < @mean + 2.5 * @std')
  error_count = (df_filtered['key_resp_trial.corr'] == 0).sum()

  error_percentage = error_count / len(df_filtered)

  mean_rt_filtered = df_filtered['key_resp_trial.rt'].mean()

  new_row_data = {"participant": df_filtered.iloc[2, df.columns.get_loc('participant')],
                   "bilingual": df_filtered.iloc[2, df.columns.get_loc('bilingual')],
                   "mean_rt": mean_rt_filtered,
                   "error_percent":error_percentage}
  new_row_df = pd.DataFrame([new_row_data])
  
  # Append the new row to the existing DataFrame
  df_all_new = pd.concat([df_all_new,new_row_df]) # Append and reset index
  # writer = pd.ExcelWriter('simon.xlsx', engine='openpyxl')  # Use openpyxl for formatting
  # writer.save()

  # print(df_all_new)

df_all_new.to_excel('simon.xlsx',index = False)

  # print(df_all_old) 
  



  # print(len(df))
  # print(len(df_filtered))
  # print(error_count)
  # print(mean_filtered)



  
  # Print details about the file (optional)
#   print(f"--- Processing: {file} ---")
```
